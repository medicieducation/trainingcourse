```python
from gensim.models import KeyedVectors

word2vec = KeyedVectors.load_word2vec_format("GoogleNews-vectors-negative300-SLIM.bin", binary=True)
```

```python
print(len(word2vec['tiger']))
print(word2vec['tiger'])
```
